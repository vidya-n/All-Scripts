```python
import asyncio
from playwright.async_api import async_playwright

async def main():
    url = 'https://www.tuftsmedicine.org/find-a-doctor?yext_retrieveFacets=true&yext_sortBys=%5B%7B%22type%22%3A%22RELEVANCE%22%7D%5D&yext_input=Melrose+Wakefield+Hospital&yext_offset=0&yext_limit=16&yext_facetFilters=%7B%7D'

    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()
        await page.goto(url)

        try:
            root1 = await page.shadow_root('tufts-yext')
            print(root1)
            root2 = await root1.query_selector('tufts-container:nth-child(2)')
            print(root2)
        except Exception as e:
            print("An error occurred:", e)
        finally:
            await browser.close()

asyncio.run(main())
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re

def convert_to_slug(name):
    # Remove any non-alphanumeric characters except spaces and hyphens
    name = re.sub(r'[^a-zA-Z0-9\s-]', '', name)
    # Convert spaces and hyphens to dashes
    name = re.sub(r'[\s-]+', '-', name)
    # Convert to lowercase
    return name.lower()

base_url = 'https://www.wellstar.org/physicians/'
url = 'https://www.wellstar.org/physicians?wellstar%2Bmedical%2Bgroup=1&v=1'
driver = webdriver.Chrome()

driver.get(url)
doctors = driver.find_elements(By.XPATH, '//div[contains(@class, "PhysicianCard_cardTitle")]')
for doc in doctors:
    # Extract text content of the element
    name = doc.text
    # Convert name to slug format
    slug_name = convert_to_slug(name)
    print(base_url + slug_name)
    
print(len(doctors))
driver.quit()
```

# Ochsner

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

url = 'https://search.ochsnerlsuhs.org/doctors?p=doctors'
driver = webdriver.Chrome()
page_count = 1
driver.get(url)
time.sleep(5)
profiles = []

while page_count <= 26:
    doctors = driver.find_elements(By.XPATH, '//h3[contains(@class, "doctor-name")]/a')

    for doc in doctors:
        #print(doc.get_attribute('href'))
        doc_url = doc.get_attribute('href')
        profiles.append(doc_url)
        
    try:
        driver.find_element(By.XPATH, '//button[contains(@aria-label, "next results page")]').click()
        time.sleep(5)
        print(f"Page {page_count} done")
        print(len(profiles))
        page_count = page_count + 1
    except Exception as e:
        print(e)
        break


driver.quit()
```
