```python
#for phy table

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import datetime
path=r'C:\Users\nyaya\Downloads\16f6c2e5-69c2-46fc-9887-f5f1d45bddd9_Export-20590c5c-744d-4529-93a9-0cf762fb2034.zip'
df = pd.read_csv(path)
num_rows_not_descoped = df[df['Status'] != 'descoped'].shape[0]
num_rows_signed_off = df[df['Status'] == 'Signed off'].shape[0]
df['due date'] = pd.to_datetime(df['Due Date'])
today = datetime.today().date()
today_timestamp = pd.Timestamp(today)
filtered_df = df[(df['due date'] < today_timestamp) & (df['Status'] != 'Signed off')]
num_unique_rows = filtered_df.shape[0]
print("PHY STATS:")
print("Total IDs to be delivered:",num_rows_not_descoped)
print("Signed off till date:",num_rows_signed_off)
print("Total number of expired IDs:",num_unique_rows)
print("Following stats of expired IDs:")
# Count the number of rows for each specified family
family_counts = filtered_df[filtered_df['GC Family'].isin(['Accesstocare', 'HCA', 'Scorpion (new)', 'Lifepoint'])].groupby('GC Family').size()

# Print the counts for each family
print("Counts for each family:")
print(family_counts)

# Calculate the total count for the specified families
total_family_count = family_counts.sum()
print("Total count for specified families:", total_family_count)

# Filter out rows with the specified families
filtered_df_excluded_families = filtered_df[~filtered_df['GC Family'].isin(['Accesstocare', 'Scorpion (new)', 'Lifepoint', 'HCA'])]

# Group the filtered DataFrame by 'Status' and count the number of rows in each group
status_counts = filtered_df_excluded_families.groupby('Status').size().reset_index(name='Count')

# Print the counts for each status
print(status_counts)

# Filter the DataFrame to select rows where 'Status' is blank
blank_status_df = filtered_df_excluded_families[filtered_df_excluded_families['Status'].isnull()]

# Count the number of rows for each researcher where 'Status' is blank
researcher_counts_blank_status = blank_status_df.groupby('Researcher').size().reset_index(name='Count')

# Print the counts for each researcher where 'Status' is blank
print("Counts for each researcher with blank status:")
print(researcher_counts_blank_status)
```
