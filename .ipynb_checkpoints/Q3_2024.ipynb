# Advent Health

```python
import requests
from bs4 import BeautifulSoup as bs

page = 0
doc_urls = set()  # Using a set to avoid duplicate URLs

while page <= 15:
    url = f'https://www.stcharleshealthcare.org/providers?search=&page={page}'
    headers = {
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
        'Accept-Encoding': 'gzip, deflate, br, zstd',
        'Accept-Language': 'en-US,en;q=0.9',
        'Sec-Cha-Ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
        'Sec-Cha-Ua-Mobile': '?0',
        'Sec-Cha-Ua-Platform': '"Windows"',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'
    }
    print("Hitting URL: ", url)
    r = requests.get(url, headers=headers)
    
    if r.status_code != 200:
        print(f"Failed to retrieve page {page}. Status code: {r.status_code}")
        break
    
    soup = bs(r.content, 'html.parser')

    # Find all 'h3' tags and extract 'a' tag links
    docs = soup.find_all('h3')
    for doc in docs:
        doc_link = doc.find('a', href=True)  # Use find() to get the first 'a' tag with 'href'
        if doc_link and doc_link['href'].strip():
            # Add only non-empty href attributes
            doc_urls.add(doc_link['href'].strip())

    print(f"Page {page} done. Total unique URLs collected so far: {len(doc_urls)}")
    page += 1

print(f"Total unique URLs collected: {len(doc_urls)}")
# If needed, print all collected URLs
# for url in doc_urls:
#     print(url)
```

#with nested request - slow processing but accurate results.

```python
import requests

offset = 1
page = 1
profiles = set()
base_find_doctor = 'https://www.sutterhealth.org/find-doctor/'
base_find_clinician = 'https://www.sutterhealth.org/find-clinician/'
while page <= 2:
    url = f'https://www.sutterhealth.org/widgets/json-doctor-results?location=&zip=&lat=&lng=&provider-gender=&medical-group=alta-bates-summit-medical-center-alta-bates-campus^^alta-bates-summit-medical-center-summit-campus&q=&online-services=&specialty=&provider-language=&health-plan=&health-plan-product=&start={offset}'

    headers = {
        'Accept': 'application/json, text/javascript, */*; q=0.01',
        'Accept-Encoding': 'gzip, deflate, br, zstd',
        'Accept-Language': 'en-US,en;q=0.9',
        'Referer': 'https://www.sutterhealth.org/doctor-search/results?location=&zip=&lat=&lng=&provider-gender=&medical-group=alta-bates-summit-medical-center-alta-bates-campus&medical-group=alta-bates-summit-medical-center-summit-campus&q=&online-services=&specialty=&provider-language=&health-plan=&health-plan-product=',
        'Sec-Cha-Ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
        'Sec-Cha-Ua-Mobile': '?0',
        'Sec-Cha-Ua-Platform': '"Windows"',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'
    }

    r = requests.get(url, headers=headers)
    docs = r.json()['payload1']['autnresponse']['responsedata']['hit']
    for i in docs:
        doc_name = i['content']['PROVIDER']['URL']['$']
        doc_url = base_find_doctor + doc_name
        try:
            rd = requests.get(doc_url, headers=headers)
            if rd.status_code == 200:
                print(doc_url)
                profiles.add(doc_url)
            else:
                doc_url = base_find_clinician + doc_name
                print(doc_url)
                profiles.add(doc_url)
        except:
            pass
    
    print(f"Page {page} done: ", len(profiles))
    offset += 10
    page += 1

print(len(profiles))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

offset = 1
page = 1
profiles = set()
driver = webdriver.Chrome()
url = 'https://www.sutterhealth.org/doctor-search/results?location=&zip=&lat=&lng=&provider-gender=&medical-group=alta-bates-summit-medical-center-alta-bates-campus^^alta-bates-summit-medical-center-summit-campus&q=&online-services=&specialty=&provider-language=&health-plan=&health-plan-product=&start=1'

driver.get(url)
time.sleep(10)

while page <= 5:
    try:
        driver.find_element(By.XPATH, '//button[text()="Accept"]').click()
    except:
        pass    
    wait = WebDriverWait(driver, 20)
    
    element = wait.until(EC.visibility_of_element_located((By.XPATH, '(//span[contains(@class, "search-count-total")])[1]')))

    docs = driver.find_elements(By.XPATH, '//h3/a')
    for doc in docs:
        profiles.add(doc.get_attribute('href'))
    print(f"Page {page} done: ", len(profiles))
    
    try:
        driver.find_element(By.XPATH, '//a[contains(., "Next")]').click()
        page += 1
        time.sleep(5)
    except Exception as e:
        print(f"Unable to click {page}. Error: ", e)
        break
    #offset += 10

print("Total doc_count:", len(profiles))
driver.quit()
```
